# ITTF Ranking

During the Tokyo Olympics 2020 which happened a few months ago, I got hooked on table tennis as I watched the Chinese table tennis team play. It was not surprising that they swept 4 out of 5 of the table tennis events held. I became curious about how the Chinese team can dominate table tennis, thus I decied to web scrape the world rankings froms the ITTF website to find more information.

```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.ittf.com/wp-content/uploads/2021/10/2021_41_SEN_MS.html"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find(id="content")
```

```python
#print(results.prettify())
```

```python
headers_list=[]
for i in results.find_all('tr', class_="tablehead"):
    title = i.text.strip()
    headers_list.append(title)
headers=headers_list[0].split("\n")
headers
```

```python
import pandas as pd
df = pd.DataFrame(columns = headers)
df
```

```python
#results.find_all('tr',class_='rrow')
```

```python
for j in results.find_all('tr',class_="rrow")[0:]:
        row_data = j.find_all('td')
        row = [tr.text for tr in row_data]
        length = len(df)
        df.loc[length] = row
df
```

```python
df2=df.copy()
list=[" ","\n","⇡1","⇣1","⇡2","⇣2","⇡3","⇡4","⇡5","⇡11","⇡17","⇡28"]
rank_list=[]
for rank in df.Rank:
    for r in range(len(list)):
        if list[r] in rank:
            rank=rank.replace(list[r],"")
    rank_list.append(rank)
df2.Rank=rank_list
df2
```

```python
df3=df2.loc[0:999]
df3
```

```python
df3.to_csv('ITTF_Ranking.csv',index=False)
```
